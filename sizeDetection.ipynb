## Use physical size of the artwork to calculate the DPI value


```python
from PIL import Image
import matplotlib.pyplot as plt

#Function that converts pixels to milimeters
def pixels_to_mm(pixels, dpi):
    return (pixels * 25.4) / dpi

#DPI is calculated using the arrow provided as a scale, 1428/7.178070866 = 198.86958859121413          HeightInPixels/LengthInInches = DPI
# 1 Pixel = 0.0050284209218914 for Height
#0.0050119346733668 - H
#0.0049978195395146 - W
# 1 Pixel = 4.284133858/854 = 0.0050165501850117 for Width


def checkLandscape_image(image_path, dpi=1427.5/7.178070866):
    img = Image.open(image_path)
    
    height_pixels = img.size[1]

    print("Size in inches",height_pixels/dpi)
    
    height_mm = pixels_to_mm(height_pixels, dpi)
    
    required_height_mm = 9  
    print("Minimum Height:",required_height_mm,"mm")
    print("Image Height:",height_mm,"mm")
    print("Pixels",height_pixels)

    plt.imshow(img)
    
    if height_mm >= required_height_mm:
        print("The image meets the required height.")
    else:
        print("The image does not meet the required height.")


def checkPortrait_image(image_path, dpi=854/4.284133858):
    img = Image.open(image_path)
    
    width_pixels = img.size[0]
    
    width_mm = pixels_to_mm(width_pixels, dpi)
    
    required_width_mm = 6
    print("Minimum Width:",required_width_mm,"mm")
    print("Image Width:",width_mm,"mm")
    print("Pixels",width_pixels)
    plt.imshow(img)
    
    if width_mm >= required_width_mm:
        print("The image meets the required width.")
    else:
        print("The image does not meet the required width.")

portrait_path = 'C:/Users/Mylu/Downloads/test/fsclogoBP.png'
landscape_path='C:/Users/Mylu/Downloads/test/fsclogoBP.png'

checkPortrait_image(portrait_path)
checkLandscape_image(landscape_path)
```

## Converts PDF to an Image 

```python
from PIL import Image
import matplotlib.pyplot as plt
import math
from pdf2image import convert_from_path

def calculate_PPI(width_pixels, height_pixels, printSize):
    # Calculate the diagonal in pixels
    diagonal_pixels = math.sqrt(width_pixels**2 + height_pixels**2)

    # Calculate Screen Diagonal (In this case the print size)
    diagonal_size_inches = math.sqrt(printSize[0]**2 + printSize[1]**2)

    print("Diagonal in pixels:", diagonal_pixels)
    print("Diagonal Size in inches:", diagonal_size_inches)

    # Calculate PPI
    ppi = diagonal_pixels / diagonal_size_inches
    print("PPI:", ppi)

    return ppi

def checkLandscape_image(image, originalArtwork_image, printSize):
    width_pixels, height_pixels = image.size
    ori_width_pixels, ori_height_pixels = originalArtwork_image.size

    # Calculate PPI based on Original Artwork Size
    ppi = calculate_PPI(ori_width_pixels, ori_height_pixels, printSize)

    # Calculate Label size in mm
    height_mm = (height_pixels * 25.4) / ppi
    min_height_mm = 9  
    print("Minimum Height:", min_height_mm, "mm")
    print("Image Height:", height_mm, "mm")
    print("Pixels:", height_pixels)

    plt.imshow(image)
    
    if height_mm >= min_height_mm:
        print("The image meets the required height.")
    else:
        print("The image does not meet the required height.")

def checkPortrait_image(image, originalArtwork_image, printSize):
    width_pixels, height_pixels = image.size
    ori_width_pixels, ori_height_pixels = originalArtwork_image.size
    print(ori_width_pixels, ori_height_pixels)

    # Calculate PPI based on Original Artwork Size
    ppi = calculate_PPI(ori_width_pixels, ori_height_pixels, printSize)

    # Calculate Label size in mm
    width_mm = (width_pixels * 25.4) / ppi
    min_width_mm = 6  
    print("Minimum Width:", min_width_mm, "mm")
    print("Image Width:", width_mm, "mm")
    print("Pixels:", width_pixels)

    plt.imshow(originalArtwork_image)
    #plt.imshow(image)
    
    if width_mm >= min_width_mm:
        print("The image meets the required width.")
    else:
        print("The image does not meet the required width.")

def process_file(file_path, originalArtwork_path, printSize):
    # Convert original artwork PDF to image
    originalArtwork_images = convert_from_path(originalArtwork_path)
    originalArtwork_image = originalArtwork_images[0]  # Process the first page

    # Check if the file is a PDF or an image
    if file_path.lower().endswith('.pdf'):
        images = convert_from_path(file_path)
        # Process the first page of the PDF
        image = images[0]
        checkLandscape_image(image, originalArtwork_image, printSize)
    else:
        image = Image.open(file_path)
        checkPortrait_image(image, originalArtwork_image, printSize)

# Defining Print Size and File Path
a4 = [8.3, 11.7]
portrait_path = 'C:/Users/Mylu/Downloads/test/fsclogoBP.png'
landscape_path = 'C:/Users/Mylu/Downloads/test/fsclogoBP.png'
originalArtwork_path = 'C:/Users/Mylu/Downloads/test/1.pdf'  # Original Artwork PDF path

# Process the files
process_file(portrait_path, originalArtwork_path, a4)
#process_file(landscape_path, originalArtwork_path, a4)
```

```python
from spire.pdf.common import *
from spire.pdf import *
import math

def calculatePPI(pxlWidth,pxlHeight,inchWidth,inchHeight):
    # Calculate the diagonal in pixels
    diagonal_pixels = math.sqrt(pxlWidth**2 + pxlHeight**2)

    # Calculate Screen Diagonal (In this case the print size)
    diagonal_size_inches = math.sqrt(inchWidth**2 + inchHeight**2)

    print("Diagonal in pixels:", diagonal_pixels)
    print("Diagonal Size in inches:", diagonal_size_inches)

    # Calculate PPI
    ppi = diagonal_pixels / diagonal_size_inches
    print("PPI:", ppi)

    return ppi

# Create a PdfDocument object
doc = PdfDocument()

# Load a PDF file
doc.LoadFromFile("C:/Users/Mylu/Downloads/test/1.pdf")

# Get the first page 
page = doc.Pages[0]

# Get the width and height of page in points
pointWidth = page.Size.Width
pointHeight = page.Size.Height

# Create a PdfUnitConvertor to convert unit
unitCvtr = PdfUnitConvertor()

# Convert units from points to inches
inchWidth = unitCvtr.ConvertUnits(pointWidth, PdfGraphicsUnit.Point, PdfGraphicsUnit.Inch)
inchHeight = unitCvtr.ConvertUnits(pointHeight, PdfGraphicsUnit.Point, PdfGraphicsUnit.Inch)

pxlWidth = unitCvtr.ConvertUnits(pointWidth, PdfGraphicsUnit.Point, PdfGraphicsUnit.Pixel)
pxlHeight = unitCvtr.ConvertUnits(pointHeight, PdfGraphicsUnit.Point, PdfGraphicsUnit.Pixel)

calculatePPI(pxlWidth,pxlHeight,inchWidth,inchHeight)
```
